```python
import pandas as pd
from sklearn.ensemble import RandomForestClassifier

data = pd.read_excel('Data/DecisionTree Exam-2.xlsx')
data.drop('ID', axis=1, inplace=True)
data.head()
```

```python
data['Pass exam'] = data['Pass exam'].map({'Yes': 1, 'No': 0})
data['Study for exam'] = data['Study for exam'].map({'Yes': 1, 'No': 0})
data['Studdent background'] = data['Studdent background'].map({'CS': 1, 'Maths': 2, 'Other': 3})
data['Working Status'] = data['Working Status'].map({'W': 1, 'NW': 0})
data
```

```python
from sklearn.tree import DecisionTreeClassifier

tree = DecisionTreeClassifier(criterion="entropy", random_state=42)
tree.fit(data.drop('Pass exam', axis=1), data['Pass exam'])

t = tree.tree_

print(t.impurity)           # entropy at each node
print(t.children_left)      # left child index
print(t.children_right)     # right child index
print(t.n_node_samples)     # samples per node
```

```python
for i in range(t.node_count):
    left = t.children_left[i]
    right = t.children_right[i]

    # node whose children are leaves
    if left != -1 and t.children_left[left] == -1 and t.children_left[right] == -1:
        # compute info gain here
        n = t.n_node_samples[i]
        n_left = t.n_node_samples[left]
        n_right = t.n_node_samples[right]

        ig = t.impurity[i] - (
            (n_left / n) * t.impurity[left] +
            (n_right / n) * t.impurity[right]
        )

        print(f"Node {i} → IG: {ig}")
```
