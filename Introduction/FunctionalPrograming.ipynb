# Factorial

```python
def factorial(x: int) -> int:
    if x == 0:
        return 1
    else:
        return x * factorial(x-1)
```

```python
factorial(4)
```

# Permutación

```python
def permutacion(n:int, r:int) -> int:
    n_fac = factorial(n)
    n_r_fac = factorial(n-r)
    return n_fac / n_r_fac
```

```python
permutacion(5, 2)
```

# Combinación

```python
def combinacion(n:int, r:int) -> int:
    per = permutacion(n, r)
    return (1 / factorial(r)) * per
```

```python
combinacion(5, 2)
```
